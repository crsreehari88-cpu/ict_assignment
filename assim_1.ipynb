```python
1.
Grocery Billing
A shopkeeper wants to calculate the total bill for two items. Write a program that takes the price and quantity of each item and prints the total bill amount.
Sample Input:
Item1 price: 45
Item1 quantity: 3
Item2 price: 30
Item2 quantity: 2
Sample Output:
Total Bill: 195
```

```python



item1_price = float(input("Item1 price: "))
item1_quantity = int(input("Item1 quantity: "))
item2_price = float(input("Item2 price: "))
item2_quantity = int(input("Item2 quantity: "))


total_bill = (item1_price * item1_quantity) + (item2_price * item2_quantity)


print("Total Bill:", total_bill)
```

Area of a Triangle (Using Heron‚Äôs Formula)
A land surveyor wants to calculate the area of an irregular triangular plot. Write a Python program to find the area of a triangle when all three sides are known, using Heron‚Äôs formula: ùë†=ùëé+ùëè+ùëê2 Area=‚àöùë†(ùë†‚àíùëé)(ùë†‚àíùëè)(ùë†‚àíùëê)
Sample Input:
Enter side a: 5
Enter side b: 6
Enter side c: 7
Sample Output:
Area of the triangle: 14.70

```python
import math

a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2

area = math.sqrt(s * (s - a) * (s - b) * (s - c))

print("Area of the triangle:", round(area, 2))
```

Paint Cost Estimator
A painter charges ‚Çπ25 per square meter. Write a program to calculate the total cost to paint a rectangular wall. The wall has length and height, and there are two windows each measuring 1.5m √ó 1.2m that should not be painted. Find the total area to be painted and its total cost.
Sample Input:
Enter wall length (in meters): 12
Enter wall height (in meters): 8
Sample Output:
Total area to be painted: 92.4 sq.m
Total painting cost: ‚Çπ2310.0


```python

length = float(input("Enter wall length (in meters): "))
height = float(input("Enter wall height (in meters): "))


wall_area = length * height


window_area = 1.5 * 1.2

total_window_area = 2 * window_area


paint_area = wall_area - total_window_area


cost = paint_area * 25

print("Total area to be painted:", paint_area, "sq.m")
print("Total painting cost: ‚Çπ" + str(cost))
```

Electricity Bill Calculator
An electricity board charges consumers based on unit consumption. The rate per unit is ‚Çπ5.2. Additionally, there is a fixed charge of ‚Çπ100 on every bill. Write a Python program that takes the number of units consumed and calculates the total bill amount.
Sample Input:
Enter number of units consumed: 250
Sample Output:
Electricity Bill: ‚Çπ1400.0

```python


units = float(input("Enter number of units consumed: "))

rate_per_unit = 5.2
fixed_charge = 100

bill = (units * rate_per_unit) + fixed_charge

print("Electricity Bill: ‚Çπ" + str(bill))
```
